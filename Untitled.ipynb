```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./dataset/the-office-lines.xlsx')
df.head()
```

```python
stopwords = set(STOPWORDS)

def cloud(speaker):
    text = df[df['speaker']==speaker].line_text.tolist()
    text = ' '.join(str(v) for v in text).lower()
 
    wordcloud = WordCloud(
        background_color='white',
        max_words=2000,
        stopwords = stopwords
    )
    
    wordcloud.generate(text)
    name = f"{speaker}'s wordcloud "
    plt.figure(figsize=(14,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(name,fontsize=15)
    plt.axis('off')
    plt.savefig('./Visualizations/'+name+'.png')
    plt.show()
```
