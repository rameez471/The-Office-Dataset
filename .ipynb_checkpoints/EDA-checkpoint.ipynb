# Explorng Deleted Scenes

```python
df['deleted'].value_counts()
```

```python
deleted = df[df['deleted']==True]
```

```python
deleted
```

```python
del_speaker = deleted.groupby('speaker').count()['scene']
del_speaker 
```

```python
plt.figure(figsize=(10,10))
wedges, texts, autotexts = del_speaker.plot.pie(label='',title='No of deleted scene by Character')

plt.show()
```
